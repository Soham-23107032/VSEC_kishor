```python
import pandas as pd
import matplotlib.pyplot as pt
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
print(iris.head())
```

```python
c=iris['species'].value_counts()
```

```python
pt.figure(figsize=(5,5))
pt.pie(c, labels=c.index, autopct='%1.1f%%', startangle=140)
pt.title('Pie chart')
pt.show()
```

```python
smean=iris.groupby('species').mean()
```

```python
colors=['grey','green', 'cyan','black']
smean.plot(kind='bar', figsize=(15,8), color=colors)
pt.title('Avg measurement of iris features by species')
pt.ylabel('species')
pt.xlabel('Avg measurement(cm)')
```
